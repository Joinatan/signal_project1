```python
def ft_analysis(x):
    fourier_coef_list = []
    for k in range(x+1):
        fourier_coef = (2/(np.pi*(1 - 4 * k**2))) * np.exp(1j * (2 * np.pi * 440) * k)
        fourier_coef_list.append(fourier_coef)
    return fourier_coef_list
```

```python
def ft_synthesis(t, n):
    fourier_coefs = ft_analysis(n)
    sig = 0
    for k in range(n):
        k2 = k - n
        sig += fourier_coefs[k] * np.exp(1j * 2 * np.pi * 440 * t * k)
    return sig

t = np.arange(0.0, 0.0025, 0.00001)
plt.title("Fourier synthesis, 9 harmonics")
plt.xlabel('t')
plt.plot(t, ft_synthesis(t, 9))
```
